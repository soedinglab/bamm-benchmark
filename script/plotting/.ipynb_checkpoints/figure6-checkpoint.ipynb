## Script for generating figure 6 in the paper for the CTCF distribution

### Since the input file is large, it will take a couple of hours to generate the figure. It requires ~20G memory.

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from mpl_stylesheet import wge_presentation, wanwan_colors
```

```python
def parse_scores(idir, suffix='.logOddsScore', reduceFold=1):
    scores = []
    for fn in os.listdir(idir):
        if fn.endswith(suffix):
            ifile = idir + fn
            if os.path.getsize(ifile) > 0:
                new_scores = pd.read_csv(ifile, sep=' ', header=None)[0].tolist()
                nums = len(new_scores)//reduceFold
                #random.shuffle(new_scores)
                scores += new_scores[:nums]
    print(len(scores))
    return scores
```

```python
def CTCF_dist(pos_scores, neg_scores, title='', save_fn='', ylim=300, colors=[ "#43a2ca","#e6550d", ], window=(0,35), figsize=(8,8)):

    # plot histogram of the log odds scores
    fig = plt.figure(figsize=figsize)
    ax1 = plt.axes()
    
    ax1.ticklabel_format(axis="y", style="sci", scilimits=(0,0), useMathText=True)
    ax1.hist([neg_scores, pos_scores], bins=60, alpha=0.8, color=colors, edgecolor = "none" )
    ax1.set_xlim(window[0],window[1])
    ax1.set_ylim(0,ylim)
    
    if title:
        ax1.set_title(title)
    ax1.set_xlabel("log odds scores")
    ax1.set_ylabel("counts")
    
    ax2 = plt.axes([0.42,0.39,0.44,0.44])
    ax2.ticklabel_format(axis="y", style="sci", scilimits=(0,0), useMathText=True)
    ax2.hist([neg_scores, pos_scores], bins=60, alpha=0.8, color=colors, edgecolor = "none" )
    
    legend=plt.legend(labels=['neg', 'pos'], loc = 'best')
    plt.show()
    
    if save_fn:
        fig.savefig(save_fn, bbox_inches='tight')
    return plt
```

```python
rfold = 100                          # fold for reducing the size of input
hmdir = '../../data/processed/CTCF/' # directory where the data is stored
suffix = '.logOddsScore'             # suffix of input files
odir_figure = "../../figures/"       # directory to store generated figures 
wge_presentation()                   # modify the figures
```

```python
neg_scores_gm = parse_scores(hmdir+"/GM/negset/", suffix=suffix, reduceFold=rfold)
pos_scores_gm = parse_scores(hmdir+"/GM/posset/", suffix=suffix, reduceFold=rfold)
```

```python
CTCF_dist(pos_scores_gm, neg_scores_gm, 
          title = '5$^{th}$ BaMM from GM12878 cells\n', 
          save_fn = odir_figure + 'CTCF_GM.pdf', 
          ylim = 7000, 
         )
```
