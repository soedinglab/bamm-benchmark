## Script for generating figure 5 in the paper

```python
from mpl_stylesheet import statPlots, wge_colors_3by5
```

```python
idir_encode = "../../data/processed/ENCODE/"
idir_selex = "../../data/processed/HTSELEXc4/"
odir_figure = "../../figures/"

sfx_time = ".time.bench"
sfx_score = ".bmscore.bench"
sfx_length = ".length.bench"

datasets_flank = [
    "bamm_k0_w8",
    "bamm_k0_w8_exd22",
    "bamm_k0_w8_exd44",
    
    "bamm_k1_w8",
    "bamm_k1_w8_exd22",
    "bamm_k1_w8_exd44",
    
    "bamm_k2_w8",
    "bamm_k2_w8_exd22",
    "bamm_k2_w8_exd44",
    
    "bamm_k5_w8",
    "bamm_k5_w8_exd22",
    "bamm_k5_w8_exd44",
]

labels_flank = [
        
    "bamm_k0_w8",
    "bamm_k0_w8_exd22",
    "bamm_k0_w8_exd44",
    
    "bamm_k1_w8",
    "bamm_k1_w8_exd22",
    "bamm_k1_w8_exd44",
    
    "bamm_k2_w8",
    "bamm_k2_w8_exd22",
    "bamm_k2_w8_exd44",
    
    "bamm_k5_w8",
    "bamm_k5_w8_exd22",
    "bamm_k5_w8_exd44",
]
```

```python
scores_flank_encode = []
scores_flank_selex = []
for data in datasets_flank:
    scores_flank_encode.append(idir_encode+data+sfx_score)
    scores_flank_selex.append(idir_selex+data+sfx_score)

cluster_names = ["$0^{th}$-order\n(0,$\pm$2, $\pm$4)", 
                 "$1^{st}$-order\n(0,$\pm$2, $\pm$4)", 
                 "$2^{nd}$-order\n(0,$\pm$2, $\pm$4)", 
                 "$5^{th}$-order\n(0,$\pm$2, $\pm$4)"]
args_flank_encode =  {
         "scores" : scores_flank_encode,
         "labels" : labels_flank,
         "nolegend": True,
         "ncluster": 4,
         "diselect": [labels_flank[9], labels_flank[11]],
         "nameclus": cluster_names,
         "colors" : wge_colors_3by5(),
         "median_colors" : wge_colors_3by5(deeper=True),
         "xlabel_box" : 'BaMMs with flanks',
         "title" : "ChIP model (n=435)",
         "ytick_scatter" : (1,2), 
         "yticklabel_scatter" : (0,1), 
         "xlabel_scatter" : 'AvRec score, $5^{th}$ BaMMs',
         "ylabel_scatter" : 'AvRec log$_{2}$(fc), $5^{th}$ BaMMs, $\pm$4bp',
}

args_flank_selex =  {
         "scores" : scores_flank_selex,
         "labels" : labels_flank,
         "nolegend": True,
         "ncluster": 4,
         "diselect": [labels_flank[9], labels_flank[11]],
         "nameclus": cluster_names,
         "colors" : wge_colors_3by5(),
         "median_colors" : wge_colors_3by5(deeper=True),
         "xlabel_box" : 'BaMMs with flanks',
         "title" : "SELEX model (n=168)",
         "ytick_scatter" : (1,2), 
         "yticklabel_scatter" : (0,1), 
         "xlabel_scatter" : 'AvRec score, $5^{th}$ BaMMs',
         "ylabel_scatter" : 'AvRec log$_{2}$(fc), $5^{th}$ BaMMs, $\pm$4bp',
}
```

```python
args_flank_encode['pstyle'] = "box"
args_flank_encode['save_fn'] = odir_figure + 'cv_encode_flank_boxplot.pdf'
args_flank_encode['noticks'] = True
statPlots(args_flank_encode)
```

```python
args_flank_encode['pstyle'] = "scatter"
args_flank_encode['save_fn'] = odir_figure + '/cv_encode_flank_scatterplot.pdf'
statPlots(args_flank_encode)
```

```python
args_flank_selex['pstyle'] = "box"
args_flank_selex['save_fn'] = odir_figure + 'cv_selex_flank_boxplot.pdf'
args_flank_selex['noticks'] = True
statPlots(args_flank_selex)
```

```python
args_flank_selex['pstyle'] = "scatter"
args_flank_selex['save_fn'] = odir_figure + 'cv_selex_flank_scatterplot.pdf'
statPlots(args_flank_selex)
```
