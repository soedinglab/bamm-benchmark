```python
from mpl_stylesheet import statPlots
```

```python
idir_encode = "../../data/processed/ENCODE/"
idir_selex = "../../data/processed/HTSELEXc4/"
odir_figure = "../../figures/"

sfx_score = ".bmscore.bench"
```

```python
datasets = [
    "bammk5_full",
    "bammk5_mask",
    "bammk1_mask",
    "inmode",
    "dichipmunk",
    "chipmunk",
    "meme",
    "cisfinder",
]

labels = [
    "BaMM($5^{th}$,full)",
    "BaMM($5^{th}$)",
    "BaMM($1^{st}$)",
    "InMoDe($2^{nd}$)",
    "diChIPMunk",
    "ChIPMunk",
    "MEME",
    "CisFinder",
]

scores_encode = []
scores_selex = []
for data in datasets:
    scores_encode.append(idir_encode+data+sfx_score)
    scores_selex.append(idir_selex+data+sfx_score)
```

```python
pstyle = 'scatter'
ytick = (0.5,1,2,4,8)
yticklabel = (-1,0,1,2,3)
```

## For ChIP-seq data from ENCODE

```python
args_encode =  {
         "scores" : scores_encode,
         "labels" : labels,
         "diselect": [labels[4], labels[0]],
         "pstyle" : pstyle,
         "nolegend" : True,
         "ytick_scatter"  : ytick,
         "yticklabel_scatter"  : yticklabel,
         "xlabel" : 'AvRec score, diChIPMunk',
         "ylabel" : 'AvRec log$_{2}$(fc), $5^{th}$ BaMM',
         "title"  : "ChIP models (n = 427)", 
         "save_fn" : odir_figure+'cv_encode_scatterplot_2.pdf',
}

statPlots(args_encode)
```

```python
args_encode['diselect'] = [labels[2], labels[0]]
args_encode['xlabel'] = 'AvRec score, $1^{st}$ BaMM'
args_encode['save_fn'] = odir_figure + 'cv_encode_scatterplot_4.pdf'
statPlots(args_encode)
```

```python
args_encode['diselect'] = [labels[6], labels[0]]
args_encode['xlabel'] = 'AvRec score, MEME'
args_encode['save_fn'] = odir_figure + 'cv_encode_scatterplot_3.pdf'
statPlots(args_encode)
```

```python
args_selex =  {
         "scores" : scores_selex,
         "labels" : labels,
         "diselect": [labels[3], labels[0]],
         "pstyle" : pstyle,
         "nolegend" : True,
         "ytick_scatter"  : ytick,
         "yticklabel_scatter"  : yticklabel,
         "xlabel" : 'AvRec score, InMoDe',
         "ylabel" : 'AvRec log$_{2}$(fc), $5^{th}$ BaMM',
         "title"  : "HT-SELEX models (n = 164)", 
         "save_fn" : odir_figure+'cv_selex_scatterplot_1.pdf',
}
statPlots(args_selex)
```
