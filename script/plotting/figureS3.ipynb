```python
from mpl_stylesheet import statPlots, wanwan_colors_add
```

```python
idir_encode = "../../data/processed/ENCODE/"
odir_figure = "../../figures/"

sfx_time = ".time.bench"
sfx_score = ".bmscore.bench"
```

```python
datasets_add = [
    "bammk5_full",
    "bammk2_full",
    "bammk2_mask",
    "bammv1_k2_new",
    "inmode",
    "dichipmunk",
    "chipmunk",
    "meme",
    "cisfinder",
]

labels_add = [
    "BaMM v2($5^{th}$)",
    "BaMM v2($2^{nd}$)",
    "BaMM v2($2^{nd}$,mask)",
    "BaMM v1($2^{nd}$)",
    "InMoDe($2^{nd}$)",
    "diChIPMunk",
    "ChIPMunk",
    "MEME",
    "CisFinder",
]

scores_add = []
times_add = []
for data in datasets_add:
    scores_add.append(idir_encode+data+sfx_score)
    times_add.append(idir_encode+data+sfx_time)
```

```python
args =  {
         "scores" : scores_add,
         "times"  : times_add,
         "labels" : labels_add, 
         "colors" : wanwan_colors_add(),
         "median_colors" : wanwan_colors_add(deeper=True),
         "pstyle" : 'dual',
         "xlog"   : True,
         "legendout": True,
         "ytick_dual": (0.5,0.6,0.7),
         "xlabel" : "Runtime (seconds)",
         "ylabel" : "AvRec score",
         "save_fn" : odir_figure + 'cv_encode_full_dualplot.pdf',
         }

statPlots(args) 
```
