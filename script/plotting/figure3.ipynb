## Script for generating figure 3 in the paper

```python
from mpl_stylesheet import statPlots
```

```python
idir_diffcell = "../../data/processed/crossCLs/"
odir_figure = "../../figures/"

sfx_time = ".time.bench"
sfx_score = ".bmscore.bench"
```

```python
datasets = [
    "bammk5_full",
    "bammk5_mask",
    "bammk1_mask",
    "inmode",
    "dichipmunk",
    "chipmunk",
    "meme",
    "cisfinder",
]

labels = [
    "BaMM($5^{th}$,full)",
    "BaMM($5^{th}$)",
    "BaMM($1^{st}$)",
    "InMoDe($2^{nd}$)",
    "diChIPMunk",
    "ChIPMunk",
    "MEME",
    "CisFinder",
]
```

```python
scores_cl = [] # cross-cell-line validation

for data in datasets:
    scores_cl.append(idir_diffcell+data+sfx_score)
    
args_cl =  {
         "scores" : scores_cl,
         "labels" : labels,
         "dotselected" : 129,
         "dotname": 'Nfkb',
         "diselect": [labels[5], labels[0]],
         "ytick_dual": (0.5,0.6,0.7),
         "ytick_scatter" : (0.5,1,2,4,8), 
         "yticklabel_scatter" : (-1,0,1,2,3), 
         "xlabel_scatter" : 'AvRec score, ChIPMunk',
         "ylabel_scatter" : 'AvRec log$_{2}$(fc), $5^{th}$ BaMM',
         "title" : "ChIP model (n=238)",
}    
```

```python
args_cl['pstyle'] = "box"
args_cl['save_fn'] = odir_figure + 'cl_boxplot.pdf'
statPlots(args_cl)
```

```python
args_cl['pstyle'] = "cumu"
args_cl['save_fn'] = odir_figure + 'cl_cumuplot.pdf'
args_cl['xtick'] = [0,100,200]
args_cl['legend'] = True
statPlots(args_cl)
```

```python
args_cl['pstyle'] = "scatter"
args_cl['save_fn'] = odir_figure + 'cl_scatterplot_1.pdf'
args_cl['legend'] = False
args_cl['xtick'] = ()
statPlots(args_cl)
```
