## Script for generating figure 2 in the paper

```python
from mpl_stylesheet import statPlots
```

```python
idir_encode = "../../data/processed/ENCODE/"
idir_selex = "../../data/processed/HTSELEXc4/"
odir_figure = "../../figures/"

sfx_time = ".time.bench"
sfx_score = ".bmscore.bench"
sfx_length = ".length.bench"
```

```python
datasets = [
    "bammk5_full",
    "bammk5_mask",
    "bammk1_mask",
    "inmode",
    "dichipmunk",
    "chipmunk",
    "meme",
    "cisfinder",
]

labels = [
    "BaMM($5^{th}$,full)",
    "BaMM($5^{th}$)",
    "BaMM($1^{st}$)",
    "InMoDe($2^{nd}$)",
    "diChIPMunk",
    "ChIPMunk",
    "MEME",
    "CisFinder",
]
```

```python
scores_encode = []
times_encode = []
for data in datasets:
    scores_encode.append(idir_encode+data+sfx_score)
    times_encode.append(idir_encode+data+sfx_time)

args_encode =  {
         "scores" : scores_encode,
         "times"  : times_encode,
         "labels" : labels,  
         "dotselected" : 128,
         "dotname": 'Elf1',
         "hx_scatter" : 0.7,
         "hy_scatter" : 0.3,
         "diselect": [labels[3], labels[0]],
         "ytick_dual": (0.5,0.6,0.7),
         "ytick_scatter" : (0.5,1,2,4,8), 
         "yticklabel_scatter" : (-1,0,1,2,3), 
         "xlabel_scatter" : 'AvRec score, InMoDe',
         "ylabel_scatter" : 'AvRec log$_{2}$(fc), $5^{th}$ BaMM',
         "title" : "ChIP model (n=427)"
}
```

```python
args_encode['pstyle'] = "scatter"
args_encode['save_fn'] = odir_figure + 'cv_encode_scatterplot_1.pdf'
statPlots(args_encode)
```

```python
args_encode['pstyle'] = "box"
args_encode['save_fn'] = odir_figure + 'cv_encode_boxplot.pdf'
statPlots(args_encode)
```

```python
args_encode['pstyle'] = "cumu"
args_encode['save_fn'] = odir_figure + 'cv_encode_cumuplot.pdf'
statPlots(args_encode)
```

```python
args_encode['pstyle'] = "dual"
args_encode['save_fn'] = odir_figure + 'cv_encode_dualplot.pdf'
statPlots(args_encode)
```

```python
scores_selex = []
times_selex = []

for data in datasets:
    scores_selex.append(idir_selex+data+sfx_score)
    times_selex.append(idir_selex+data+sfx_time)
    
args_selex =  {
         "scores" : scores_selex,
         "times"  : times_selex,
         "labels" : labels,
         "dotselected" : 36,
         "dotname": 'ERG',
         "hx_scatter" : 0.7,
         "hy_scatter" : 0.3,
         "diselect": [labels[6], labels[0]],
         "ytick_dual": (0.3,0.4,0.5,0.6),
         "xlim_dual" : (2,90000),
         "ytick_scatter" : (0.5,1,2,4,8), 
         "yticklabel_scatter" : (-1,0,1,2,3), 
         "xlabel_scatter" : 'AvRec score, MEME',
         "ylabel_scatter" : 'AvRec log$_{2}$(fc), $5^{th}$ BaMM',
}
```
