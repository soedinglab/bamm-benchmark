```python
from mpl_stylesheet import statPlots, wge_colors_4by3
```

```python
idir_supp = "../../data/processed/ENCODE/supp/"
odir_figure = "../../figures/"

sfx_time = ".time.bench"
sfx_score = ".bmscore.bench"
```

```python
datasets = [
    "pengbamm_k0_w8_exd22",
    "pengbamm_k1_w8_exd22",
    "pengbamm_k2_w8_exd22",
    "pengbamm_k5_w8_exd22",
    "pengbamm_k0_w12_exd22",
    "pengbamm_k1_w12_exd22",
    "pengbamm_k2_w12_exd22",
    "pengbamm_k5_w12_exd22",
    "pengbamm_k0_w8_exd22_full_shootpeng",
    "pengbamm_k1_w8_exd22_full_shootpeng",
    "pengbamm_k2_w8_exd22_full_shootpeng",
    "pengbamm_k5_w8_exd22_full_shootpeng",
]

labels = [
     "w8, k=0, mask",
     "w8, k=1, mask",
     "w8, k=2, mask",
     "w8, k=5, mask",
     "w12, k=0, mask",
     "w12, k=1, mask",
     "w12, k=2, mask",
     "w12, k=5, mask",
     "w8, k=0, full",
     "w8, k=1, full",
     "w8, k=2, full",
     "w8, k=5, full",
]

cluster_names = [
    "core 8 bp\nmasking",
    "core 12 bp\nmasking",
    "core 8 bp\nfull set"
]
```

```python
scores = []
times = []
for data in datasets:
    scores.append(idir_supp+data+".best"+sfx_score)
    times.append(idir_supp+data+sfx_time)
```

```python
args_flank =  {
         "scores" : scores,
         "times"  : times,
         "labels" : labels,
         "pstyle" : 'box',
         "nolegend": True,
         "ncluster": 3,
         "nameclus": cluster_names,
         "colors" : wge_colors_4by3(),
         "median_colors" : wge_colors_4by3(deeper=True),
         "xlabel_box" : 'BaMMs with flanks',
         "title" : "ChIP model (n=435)",
         "ytick_scatter" : (1,2,4), 
         "yticklabel_scatter" : (0,1,2), 
         "xlabel_scatter" : 'AvRec score, $5^{th}$ BaMMs, masking',
         "ylabel_scatter" : 'AvRec log$_{2}$(fc), $5^{th}$ BaMMs, full set',
         "save_fn" : odir_figure + 'fullEM_vs_masking_encode1.pdf'
}

statPlots(args_flank)
```

```python
args_flank["diselect"] = [labels[3], labels[11]]

args_flank["pstyle"] = "scatter"
args_flank["save_fn"] = odir_figure + 'fullEM_vs_masking_encode2.pdf'
statPlots(args_flank)
```
