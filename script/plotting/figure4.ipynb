## Script for generating figure 4 in the paper

```python
from mpl_stylesheet import statPlots
```

```python
idir_selexchip = "../../data/processed/SELEXonCHIP/"
idir_chipselex = "../../data/processed/CHIPonSELEX/"
odir_figure = "../../figures/"

sfx_time = ".time.bench"
sfx_score = ".bmscore.bench"
sfx_length = ".length.bench"

datasets = [
    "bammk5_full",
    "bammk5_mask",
    "bammk1_mask",
    "inmode",
    "dichipmunk",
    "chipmunk",
    "meme",
    "cisfinder",
]

labels = [
    "BaMM($5^{th}$,full)",
    "BaMM($5^{th}$)",
    "BaMM($1^{st}$)",
    "InMoDe($2^{nd}$)",
    "diChIPMunk",
    "ChIPMunk",
    "MEME",
    "CisFinder",
]
```

```python
selected = ['ELF1', 'CEBPB', 'POU2F2', 'FOXA2']

scores_sc = [] # train on chip and test on selex
scores_cs = [] # train on selex and test on chip

for data in datasets:
    scores_sc.append(idir_selexchip+data+sfx_score)
    scores_cs.append(idir_chipselex+data+sfx_score)
      

dot = 10
dotname = 'CEBPB'

args_sc = {
         "scores" : scores_sc,
         "labels" : labels,
         "xcutoff": 0.05,
         "selected": selected,
         "dotselected" : dot,
         "dotname": dotname,
         "hx_scatter" : 0.7,
         "hy_scatter" : 0.3,
         "ytick"  : (0.1,0.3,0.5,0.7),
         "title"  : "SELEX model (n=16)",
         "ylabel" : "ChIP AvRec score",
}

args_cs = {
         "scores" : scores_cs,
         "labels" : labels,
         "xcutoff": 0.05,
         "selected": selected,
         "dotselected" : dot,
         "dotname": dotname,
         "hx_scatter" : 0.7,
         "hy_scatter" : 0.3,
         "ytick"  : (0.1,0.3,0.5,0.7),
         "title"  : "ChIP model (n=16)",
         "ylabel" : "SELEX AvRec score",
}
```

```python
args_cs['pstyle'] = "box"
args_cs['save_fn'] = odir_figure + 'cs_boxplot.pdf'
statPlots(args_cs)
```

```python
args_cs['pstyle'] = "cumu"
args_cs['save_fn'] = odir_figure + 'cs_cumuplot.pdf'
args_cs['xtick'] = (0,5,10,15)
args_cs['legend'] = True 
statPlots(args_cs)
```

```python
args_sc['pstyle'] = "box"
args_sc['save_fn'] = odir_figure + 'sc_boxplot.pdf'
statPlots(args_sc)
```

```python
args_sc['pstyle'] = "cumu"
args_sc['save_fn'] = odir_figure + 'sc_cumuplot.pdf'
args_sc['xtick'] = (0,5,10,15)
statPlots(args_sc)
```
